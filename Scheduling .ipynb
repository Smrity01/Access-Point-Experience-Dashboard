```python
import Twitter
import Youtube
from Merge_Script import merging
import time
```

```python
def thread_Youtube():
# Objective : To run Youtube.py 
    Youtube.main() 
```

```python
def thread_Twitter():
# Objective : To run Twitter.py  
    Twitter.main() 
        
```

```python
def thread_Merge():
# Objective : To run Merging.py
    merging()
        
```

```python


import threading
if __name__ == "__main__": 
    '''
        Objective: To run Twitter and Youtube script parallely and then run Merging script after an interval of 5 mins
    '''
    while True:
        t1 = threading.Thread(target = thread_Youtube)
        t2 = threading.Thread(target = thread_Twitter)
        t3 = threading.Thread(target = thread_Merge)
        try:
            t1.start()
            t2.start()
            t1.join()
            t2.join()
            time.sleep(5*60)
            t3.start()
            t3.join()
            #break
        except thread.error as error:
             print("Error : " + str(error))
```
